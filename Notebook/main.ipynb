```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\MOHIT JHA\Desktop\Project 1\Data\train.csv.zip")
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.isna().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
```

```python
df['Order Date'].isnull().sum()
```

```python
df['Revenue'] = df['Sales']
```

```python
print(df.columns)
```

Sales by Category

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

Sales by Region

```python
df.groupby('Region')['Sales'].sum()
```

Monthly Trend

```python
df['Month'] = df['Order Date'].dt.month
df.groupby('Month')['Sales'].sum()
```

Top Products

```python
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

Export Clean Data
